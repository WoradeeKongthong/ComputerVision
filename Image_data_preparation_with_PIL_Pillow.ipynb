```
import PIL
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

# Load and display image using Pillow.

```
# load the image
img = PIL.Image.open('cake.jpg')

# summarize image details
print(type(img))
print(img.format)
print(img.mode)
print(img.size)

# show the image
# image.show()
plt.imshow(img)
plt.show()
```

# Load image as numpy array using matplotlib.image

```
# load image as pixel array
pixels = matplotlib.image.imread('cake.jpg')

# summarize pixels details
print(type(pixels))
print(pixels.dtype)
print(pixels.shape)

# display the pixel array as an image
plt.imshow(pixels)
plt.show()
```

# Convert Pillow Image to and from numpy array

```
img = PIL.Image.open('cake.jpg')
print('Original image')
plt.imshow(img)
plt.show()

# convert PIL image to numpy array
img_array = np.asarray(img)

# summarize array details
print('\nAfter conversion')
print('type : ', type(img_array))
print('image array shape :', img_array.shape)
```

```
# Convert back to Pillow image
img2 = PIL.Image.fromarray(img_array)

# display the image
print('img2')
plt.imshow(img2)
plt.show()

# summarize image details
print(img2.format)
print(img2.mode)
print(img2.size)
```

# Resize image : thumbnail vs resize

```
# thumbnail all images to have the same width or height

# load image
img = PIL.Image.open('Homu/Homu3.jpg')
print('original image size : ',img.size)
plt.imshow(img)
plt.show()

# create a thumbnail and preserve aspect ratio
img.thumbnail((200,200))
print('after thumbnail : ', img.size)
plt.imshow(img)
plt.show()
```

```
# resize : force into a new shape

# load image
img = PIL.Image.open('Homu/Homu3.jpg')
print('original image size : ',img.size)
plt.imshow(img)
plt.show()

# resize
img_resized = img.resize((200,200))
print('after resize : ', img_resized.size)
plt.imshow(img_resized)
plt.show()
```

```
# load image
img = PIL.Image.open('cake.jpg')

plt.subplot(131)
plt.imshow(img)

# rotate 45 degrees not expand
plt.subplot(132)
plt.imshow(img.rotate(45))

# rotat 45 degrees with expand
plt.subplot(133)
plt.imshow(img.rotate(45, expand=1, fillcolor='white'))

plt.show()
```

```
# Crop

# load image
img = PIL.Image.open('Homu/Homu3.jpg')
print('original size : ',img.size)
#plot
plt.subplot(121)
plt.imshow(img)

#cropped image
cropped = img.crop((1000,250,2000,1000))
# plot 
plt.subplot(122)
plt.imshow(cropped)
plt.show()
```
